```python
#elif means else if statement. it is always used after the if statement to test the scond condition.
# esle statement will only be executed if any of the two statements were not executed due to not satisfying the conditions
a=3
b=7
```

```python
if a==6:
    print("Welcome to conditional statements by dr Sibiya")
elif b==7:
    print(""""This was printed because the condition of the if statement was not true, 
          otherwise it would have been the one printed instead""")
    
else:
    print("The value in the a variable is not equal to 2")
```
