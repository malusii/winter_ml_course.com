# Loops

```python
# A for loop loop used with a range function
for number in range(5):
    print(number)
```

```python
# A for loop used to return the elemnets of an array list
List_array = ["Cars", 1, "People", 2.45,"youth", "learning" ]
```

```python
# Now use a for loop to loop within the List_array while print the elements of the array
```

```python
for data in List_array:
    print(data)
```

```python
# Now use a for loop to iterate in string letters
```

```python
name = "Kevin"
```

```python
for letter in name:
    print(letter)
```

# While loop

```python
# The while loop continues to loop as long as the condition is True
# The while loop condition needs to be updated until the iterations are over
```

```python
dat_number = 0
```

```python
while dat_number <= 15:
    print(dat_number)
    dat_number += 1
```

```python
# Now use the while loop to return the letters of a string
surname = "Smith"
dat_number2 = 0
```

```python
while dat_number2 < 5:
    print(surname[dat_number2])
    dat_number2 += 1
    
```

# Break and Continue key words in loops

```python
dat_number3 = 0
```

```python
# These are used within loops
# Below is a code for demonstrating a break
while dat_number3 <= 15:
    if dat_number3 == 10:
        break
    print(dat_number3)
    dat_number3 += 1
```

```python
# Now let us use the continue loop
# Below is a code for demonstrating a break
for number in range(15):
    if number== 2 or number== 4 or number == 7:
        continue
    print(number)
    
    
```
