# Type Casting


* Type casting is all about transforming one data type to another data type. This also includes the capability of specifying the data type in a variable

```python
# Undersatanding dataype transformation
data_1 = "23" # The information stored in a variable called data_1 is a string since it is enclosed within the qoutes
              # At first you may think that it is an integer since you may have seen 23. 
              #However, the 23 is no more an integer number once it is enclosed within single or double qoutes, instead it becomes a string.
        
```

```python
# Let us check the data type in data_1 variable
type(data_1)
```

```python
# str which results after checking the data type in data_1 means that the data type is a string
```

```python
# Let us see what we get if we multiply the contents of data_1 by 2
data_1 * 2
```

```python
# if you choose to use the print function to display the result
print(data_1*2)
```

```python
# These results show that if you multiply the string by a number the result will be a repeated string by a number of times used
```

```python
# What if we wanted to transform the contents of data_1 to an integer number 23?
data_1 = int(data_1)# This transforms the contents of data_1 variable from string to an integer
```

```python
# Let us now check the data type in data_1 after the transformation has been made
type(data_1)
```

```python
# int which results simply means that the data type inside of the data_1 variable is now an integer
# We can now multiply the contents of data_1 variable by 2 just like we did before to see the result
data_1*2
```

```python
# if you choose to use the print function to display the result
print(data_1*2)
```

# Logical operators

* Logical operators always return True or False when used

```python
# Equality operator

1==1 # This returns True because 1 is equal to 1
```

```python
1==2 # This returns False because 1 is not equal to 2
```

```python
# Not Equal operator
1!=1 # This returns False because 1 is equal to 1 and here the oprator used says 1 is not equal to 1 
```

```python
6!=5 # The Not Equal operator has been used correctly because 6 is not equal to 5, so True is returned
```

```python
# Less than operator
1<2 # True is returned because 1 is less than 2
```

```python
3<2# False is returned because 3 is not less than 2
```

```python
# Less than or Equal operator
3 <=3 # The equal sign in this compound assignment is the one that True to be returned  
```

```python
# Greater than or Equal operator
6 >=9 # False is returned because the none of the two signs in the compound logical operator is true 
```
