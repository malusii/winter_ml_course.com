# Understanding class variables and instance variables

```python
class Cities:
    # These are class variables and can be shared by all object instances of the given class.
    #These variables are owned by clasess
    Province_1 = "Gauteng"
    Province_2 = "Free State"
    Province_3 = "KZN"
    
    # The following are class variables that are private
    __Bloemtemperature = 28
    _Jozitemperature = 15
    
    def __init__(self,*Town):
        #These are instance varibles and are by the given instance of the class 
        # The values of these variables can be different depending on the the instances
        self.Town = Town
    def Read_province_temp(self):
        print("The temperatures in the {} province will be around 19 degrees celcius and {}".format(self.Province_2,self.Town[0]))
        
    def product_of_temperatures(self):
        prod_temp = self.__Bloemtemperature*self._Jozitemperature
        return prod_temp
        
        
```

```python
p1 = Cities("Bloemfontein is partly warm","Joburg is cold","Durban is warm" )
p2 = Cities("Welkom is warmer")
p3 = Cities()
```

```python
p1.Read_province_temp()
p2.Read_province_temp()
```
