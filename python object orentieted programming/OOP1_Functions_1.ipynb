# Functions with positional arguments

```python
def func_pos_arg(number_1,number_2):
    print(number_1,number_2)
    
```

```python
# call a function with positional arguments
func_pos_arg(2,3)
```

```python
# now swap the arguments to further understand functions with positional arguments
func_pos_arg(3,2)
```

## Functions with positional arguments by using *


```python
# This function is used when the number of positonal arguments to be used is not known
# The arguments are then stored as tuple and can be accessed by means of index
```

```python
def func_pos_arg2(*Numbers):
    print(Numbers[3])
    print(Numbers[0])
    print(Numbers[1])
    print(Numbers[2])
```

```python
func_pos_arg2(1,2,3,4)
```

# Functions with keyword arguments

```python
def keywordrg_func(name_1,name_2):
    print(name_1,name_2)
```

```python
keywordrg_func(name_1="John",name_2="Peter")
```

```python
#now swap the arguments to better understand the functions with key arguments
keywordrg_func(name_2="Peter",name_1="John")
```

It can be seen that in functions with keyword arguments, the position of the arguments does not matter

## Functions with keyword arguments byusing **

```python
#This function is used when the number of keyword arguments to be used is not known
# The arguments are then stored as a dictionary and can be accessed by means of key
```

```python
def keywordrg_func2(**Names):
    print(Names["name_1"])
    print(Names["name_3"])
    print(Names["name_4"])
    print(Names["name_2"])
    
```

```python
keywordrg_func2(name_2="Peter",name_1="John",name_4="Mercy",name_3="Kevin")
```
