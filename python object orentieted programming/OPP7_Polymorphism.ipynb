```python
class lingo:   # Abstract class
    def say_hello(self):
        print("This is an abstract class")
        #raise NotImplementedError ("The function say_hello better be used in the child class")
        # The abstract class is just there to make sure that the objects use the same interface
```

```python
class Sesotho(lingo):
    def say_hello(self):
        print("Dumela")
    #pass
```

```python
class Afrikaans(lingo):
    def say_hello(self):
        print("Goeie dag")
```

```python
def language(lang):    # creating achannnel/interface that differnt types of objects will use
    lang.say_hello()
```

```python
Khotso = Sesotho()
```

```python
Pieter = Afrikaans()
```

```python
language(Khotso)
```

```python
language(Pieter)
```
