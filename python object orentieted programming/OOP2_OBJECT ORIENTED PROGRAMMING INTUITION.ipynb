# Encapsulation in classes

* Encapsulation is a fundamental feature of OOP where we are able to group certain variables and functions within a class

```python
# Let us create a class and demonstrate the concept of encapsulation
```

```python
class Person:      # Here we are creating a class/blue print
    name = "Malusi" # here we are creating a first variable within a class- This encapsulation
    surname = "Sibiya" # here we are creating a second variable within a class- This encapsulation
    
    
    def age_calculator(self):
        self.age = int(input("Enter your age "))
        print("{} {} you are {} years old ".format(self.name,self.surname,self.age))
```

```python
creature = Person() # We are instantiating an object. This means we are making an access to  variables and functions of a 
                    #class by using the object creature
```

```python
# Variables of a class are often known as class attributes
# By using an instantiated object, we can access the variables/attributes of a class
```

```python
# Let us access and print the contents of the class variables/attributes
creature.name
```

```python
# However since we are using Jupyter IDE we are able to access and dispaly the contents without using a print function
# Let us now use a print function
print(creature.name)
```

```python
# We can also acess the contents of a varible/attribute called surname
print(creature.surname)
```

```python
# Remember, the object cal also access or call the created function within a class
# Functions within classes are called methods
creature.age_calculator()
```
