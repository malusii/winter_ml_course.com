# Inheritance

```python
# Let us now see how inheritance can be created between two classes
```

```python
class Sequences:    #Creating a parent class
    number_1 = 0
    sequence_1 = number_1 + 1
    
    def addition(self):
        print((self.sequence_1)*3)
        
```

```python
class Logics(Sequences): # creating the child class
    step_1 = "The first step to be taken"
    
    
```

```python
p_1 = Logics() # Instantiating the object of the child class
```

```python
p_1.sequence_1 # we have access to the contents of the parent class
```

```python
p_1.addition() # we have access to the contents of the parent class
```

```python
p_1.step_1 # we also have access to the contents of the child class
```
